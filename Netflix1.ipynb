```python
import os
import pandas as pd
os.chdir("F:\Data Science FUN with lots of exercicss and tools\DATA WORLD cool datasets")
netflix=pd.read_csv("netflix.csv")
netflix.head()
```

```python
netflix = netflix.dropna(axis=0)
netflix.head()
```

```python
netflix.info()
```

```python
## Objective: Do Netflix users prefer older movies to newer ones?
## Step 1: Segregate 'Netflix' by release year
netflix_by_years = netflix.groupby("release year")
```

```python
netflix_by_years.describe(
)
```

```python
netflix_by_years[["user rating score"]]
```

```python
## Step 2: Calculate median of user rating scores for each year
median_ratings_by_year = netflix_by_years[["user rating score"]].median()
median_ratings_by_year.head()
```

```python
## Step 3: Plot median ratings vs. release year
plt.scatter(x=median_ratings_by_year.index, y=median_ratings_by_year)
plt.xlabel = "release year"
plt.ylabel = "median user rating score"
plt.show(
)
```
